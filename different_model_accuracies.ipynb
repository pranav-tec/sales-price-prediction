```python
import pandas as pd
import numpy as np
df=pd.read_csv('bpp_training_data.csv')
df.head(100)
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True)
```

```python
#seperating output variable
x=df.drop(['SalePrice'],axis=1)
x
```

```python
#assingning output variable
y=df.SalePrice
y
```

```python
from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X_train_scaled = ss.fit_transform(x_train)
X_test_scaled = ss.transform(x_test)
```

```python
#model building
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
lr_train_mse = sqrt(mean_squared_error(y_train, y_train_pred))
lr_train_r2 = r2_score(y_train, y_train_pred)
lr_test_mse = sqrt(mean_squared_error(y_test, y_test_pred))
lr_test_r2 = r2_score(y_test, y_test_pred)
```

```python
print(lr_train_mse)
```

```python
print(lr_test_mse)
```

```python
print(lr_test_r2)
```
